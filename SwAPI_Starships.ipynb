```python
import requests
import json
from pprint import pp
```

Get the responce for all the pages and save it in the list.

```python
url="https://swapi.dev/api/starships/?page=1"
# Results_list=[]
result=[]
while url:
    response=requests.get(url)
    data=response.json()
    # Results_list.append(data)
    result.extend(data['results'])
    url=data['next']
pp(result)
```

1. Create a function for finding the cargo capacity which is 'unknown' and the names starts with 'Naboo'and return a list with their name,model and pilots.

```python
def return_data(n):
    result_data=[]
    for data in result:
        if data['cargo_capacity']=='unknown' and data['name'].startswith(n):
            result_data.append([data['name'],data['model'],data['pilots']])
    return result_data
print(return_data('Naboo'))
```

2. Write a function to find the starship_class with no passengers.

```python
def return_starship_class():
    starship_result=[]
    for data in result:
        if data['passengers']=='0':
            starship_result.append(data['starship_class'])
    return starship_result
print(return_starship_class())
```

 3. Find length and crew with with 1 pilot and 1 film 

```python
lenght_crew=[]
for data in result:
    if len(data['pilots'])==1 and len(data['films'])==1:
        lenght_crew.append([data['length'],data['crew']])
print(lenght_crew)
```

4. Find minimum and maximum of max-atmosphering-speed where the starship-class is starfighter and put them with manufature,length,and crew in 2 dictionaries. one for min_speed and other dict for max_speed.

```python
max_atmos_speed={}
max_atmos_list=[]
min_atmos_speed={}
for data in result:
    if data['starship_class']=='starfighter':
        max_atmos_list.append(data['max_atmosphering_speed'])
minimum_speed=min(max_atmos_list)
maximum_speed=max(max_atmos_list)
for data in result:
    if data['starship_class']=='starfighter' and data['max_atmosphering_speed']==maximum_speed:
        max_atmos_speed['Manufacturer_Name']=data['manufacturer']
        max_atmos_speed['Length_size']=data['length']
        max_atmos_speed['Crew-size']=data['crew']
        max_atmos_speed['Max_Speed']=maximum_speed
    if data['starship_class']=='starfighter' and data['max_atmosphering_speed']==minimum_speed:
        min_atmos_speed['Manufacturer_Name']=data['manufacturer']
        min_atmos_speed['Length_size']=data['length']
        min_atmos_speed['Crew-size']=data['crew']
        min_atmos_speed['Min_Speed']=minimum_speed
# print(minimum_speed,maximum_speed)
print("****************Maximum**************")
print(max_atmos_speed)
print("****************Minimum**************")
print(min_atmos_speed)
```

5. for given model , find the hyperdrive_rating and 'MGLT' and return them in dictionary. 

```python
def user_data(model):
    hyper_Mglt={}
    for data in result:
        # model=int(input("enteryour"))
        if data['model']==model:
            hyper_Mglt["Hyperdrive_rating"]=data['hyperdrive_rating']
            hyper_Mglt['MGLT']=data['MGLT']
    return hyper_Mglt
print(user_data("J-type star skiff"))


```
