1. Create a list that stores a certain number of items to work off of. We have the range of item ids from the MAX ID endpoint. I suggest that you start small and expand the dataset

```python
import requests
import json
from pprint import pp
```

```python
url="https://hacker-news.firebaseio.com/v0/item/{}.json?print=pretty"
results=[]
for i in range(1,1000): #get 999 user id's from the API
    results.append(requests.get(url.format(str(i))).json()) #put them into list
pp(results)
```

```python
def return_user(user):
    usr_ct=0
    for usr in results:
        if usr['by']==user:
            usr_ct+=1
            pp(usr)
    return f"Total count:{usr_ct}"
    
    
print(return_user('pg'))
```

12. Write a function to see which hour of the day is the most common for a user to post an item

```python
def time_hour():
    from collections import Counter
    ct_hour=[]
    for item in results:
        t=datetime.fromtimestamp(item['time'])
        ct_hour.append(t.hour)
    mst_commn=Counter(ct_hour)  # print(t.hour)
    for tm in mst_commn.most_common(1):
        if tm[0] >12:
            return f"Most common hour for a user:{tm[0]-12}Pm" 
    
print(time_hour())
```
